```python
df_games = pd.read_csv(
    csv_file,
    on_bad_lines='skip',
    encoding='utf-8',
    low_memory=False
)

print(f"Loaded: {len(df_games):,} rows, {len(df_games.columns)} columns")

df_games.reset_index(inplace=True)
df_games.rename(columns={'index': 'AppID'}, inplace=True)
df_games.rename(columns={'AppID': 'AppID_old'}, inplace=True)

old_cols = df_games.columns.tolist()[1:] 

correct_column_names = [
    'AppID',  
    'Name', 'Release date', 'Estimated owners', 'Peak CCU', 'Required age',
    'Price', 'DiscountDLC count', 'About the game', 'Supported languages',
    'Full audio languages', 'Reviews', 'Header image', 'Website', 'Support url',
    'Support email', 'Windows', 'Mac', 'Linux', 'Metacritic score',
    'Metacritic url', 'User score', 'Positive', 'Negative', 'Score rank',
    'Achievements', 'Recommendations', 'Notes', 'Average playtime forever',
    'Average playtime two weeks', 'Median playtime forever',
    'Median playtime two weeks', 'Developers', 'Publishers', 'Categories',
    'Genres', 'Tags', 'Screenshots', 'Movies'
]


df_games = df_games.iloc[:, :len(correct_column_names)]
df_games.columns = correct_column_names



# Verify
print(f"\n1. First AppID: {df_games['AppID'].iloc[0]}")
print(f"2. First Name: {df_games['Name'].iloc[0]}")
print(f"3. First Release date: {df_games['Release date'].iloc[0]}")
print(f"4. First Genres: {df_games['Genres'].iloc[0]}")
print(f"5. First Price: {df_games['Price'].iloc[0]}")

print("\n6. Sample data:")
print(df_games[['AppID', 'Name', 'Release date', 'Genres', 'Price']].head(5))
```

```python
import shutil
from pathlib import Path

project_data = Path("/home/boson/teams/z2/data/steam/raw")
project_data.mkdir(parents=True, exist_ok=True)

shutil.copy(csv_file, project_data / "steam_games_metadata_original.csv")

df_games.to_csv(
    project_data / "steam_games_metadata.csv",
    index=False,
    encoding='utf-8'
)
df_clean = df_games.copy()

for col in df_clean.select_dtypes(include=['object']).columns:
    df_clean[col] = df_clean[col].apply(
        lambda x: x.encode('utf-8', errors='ignore').decode('utf-8') 
                  if isinstance(x, str) else x
    )

df_clean.to_parquet(
    project_data / "steam_games_metadata.parquet",
    compression='gzip',
    index=False
)

import os
csv_size = os.path.getsize(project_data / "steam_games_metadata.csv") / 1024**2
parquet_size = os.path.getsize(project_data / "steam_games_metadata.parquet") / 1024**2

print(f"Location: {project_data}")
print(f"  - CSV: {csv_size:.1f} MB")
print(f"  - Parquet: {parquet_size:.1f} MB")
print(f"  - Rows: {len(df_games):,}")
print(f"  - Columns: {len(df_games.columns)}")
```

```python
test_df = pd.read_parquet(project_data / "steam_games_metadata.parquet")
print(f"Successfully read parquet: {len(test_df):,} rows")
```

```python
def parse_genres(genre_str):
    if pd.isna(genre_str) or genre_str == '' or str(genre_str) == 'nan':
        return []
    return [g.strip() for g in str(genre_str).split(',') if g.strip()]
df_games['Game_Genres'] = df_games['Tags'].apply(parse_genres)
df_games['steam_features'] = df_games['Genres'].apply(parse_genres)

for idx in range(10):
    if len(df_games.iloc[idx]['Game_Genres']) > 0:
        print(f"\n{df_games.iloc[idx]['Name'][:45]:45} → {df_games.iloc[idx]['Game_Genres']}")


all_game_genres = [g for sublist in df_games['Game_Genres'] for g in sublist]
game_genre_counts = pd.Series(all_game_genres).value_counts()

print(f"Total Unique Game Genres: {len(game_genre_counts)}")
print("Top 20 Game Genres:")
print(game_genre_counts.head(20))

has_game_genre = df_games['Game_Genres'].apply(len) > 0
print(f"Games with Game Genres: {has_game_genre.sum():,} ({has_game_genre.mean()*100:.1f}%)")

df_games.to_parquet(
    project_data / "steam_games_with_parsed_genres.parquet",
    compression='gzip',
    index=False
)


print(f"   Location: {project_data / 'steam_games_with_parsed_genres.parquet'}")
```
